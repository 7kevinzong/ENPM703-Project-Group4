```python
import yfinance as yf
import numpy as np
```

```python
# download training data

ticker = 'CL=F'
data = yf.download(ticker, start='2024-01-01', end='2025-01-01')

# Clean data
data = data.dropna()  # Drop rows with any NaN
data = data[~data.index.duplicated()]  # Drop duplicate indices

# Calculate log returns
data['log_return'] = np.log(data['Close'] / data['Close'].shift(1))

data = data.dropna()  # Drop first row with NaN return

# Save the cleaned data to a CSV file
data.to_csv('CL_F_data.csv', index=True)  # 'index=True' to include the date index in the CSV file
print("Data has been saved to CL_F_data.csv")
```
